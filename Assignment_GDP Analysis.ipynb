### >>>Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

```python
# Getting the item names
df_states.Item
```

```python
#Selecting Required Rows and renaming the column '2014-15' to GSDP for better visualisation.

df_GSDP = df_states.loc[(df_states.Item == "Gross State Domestic Product")][['2014-15','State']].rename(columns={'2014-15':'GSDP'})



df_GSDP.head()
```

```python
#Selecting the primary sector

df_primary =df_states.loc[(df_states.Item == "Primary")][['2014-15','State']].rename(columns = {'2014-15':'Primary_GSVA'}) 

df_primary.head()
```

```python
#Selecting the secondary sector

df_secondary =df_states.loc[(df_states.Item == "Secondary")][['2014-15','State']].rename(columns = {'2014-15':'Secondary_GSVA'}) 

df_secondary.head()
```

```python
#Selecting the tertiary sector

df_tertiary =df_states.loc[(df_states.Item == "Tertiary")][['2014-15','State']].rename(columns = {'2014-15':'Tertiary_GSVA'}) 

df_tertiary.head()
```

```python
# Merging all 3 sectors on State column

df_sectors = pd.merge(df_primary,df_secondary, how = 'inner', on = 'State').merge(df_tertiary,how= 'inner',on = 'State')

df_sectors.head()
```

```python
# Merging the sectors dtaframe with the GSDP dataframe to plot the contribution

df_GSDP_sectors = pd.merge(df_GSDP,df_sectors, how = 'inner', on = 'State')

df_GSDP_sectors.head()
```

```python
# Calculating the % contribution and creating respective columns.

df_GSDP_sectors['Primary_Contribution_(%)'] = (df_GSDP_sectors['Primary_GSVA']/df_GSDP_sectors['GSDP'])*100

df_GSDP_sectors['Secondary_Contribution_(%)'] = (df_GSDP_sectors['Secondary_GSVA']/df_GSDP_sectors['GSDP'])*100

df_GSDP_sectors['Tertiary_Contribution_(%)'] = (df_GSDP_sectors['Tertiary_GSVA']/df_GSDP_sectors['GSDP'])*100

df_GSDP_sectors.head()
```

```python
# Filtering only the % contribution columns per state to plot the graph

plot_GSDP_sectors = df_GSDP_sectors.filter(['State','Primary_Contribution_(%)','Secondary_Contribution_(%)','Tertiary_Contribution_(%)'])

#plot_GSDP_sectors = plot_GSDP_sectors.set_index('State').iloc[:]

plot_GSDP_sectors.head(50)
```

```python
#Plotting the contribution of three sectors.

fig, ax1 = plt.subplots(figsize=(18,8))

tidy = plot_GSDP_sectors.melt(id_vars = 'State').rename(columns=str.title)

#sns.barplot(x ='State',y = 'Value', hue = 'Variable',data = tidy)

barplot_4 = sns.barplot(x ='State',y = 'Value', hue = 'Variable',data = tidy)

plt.xlabel('States',fontname="Arial",fontsize='15')

plt.ylabel('Contribution to GSDP (%)',fontname="Arial",fontsize='15')

plt.title('Contribution of various sectors to State GDP',fontname="Arial",fontweight="bold",fontsize='15')

plt.xticks(rotation = 90)

plt.show()
```

### >>>Find the top 3/4/5 sub-sectors (such as agriculture, forestry and fishing, crops, manufacturing etc., not primary, secondary and tertiary) that contribute to approximately 80% of the GSDP of each category.

Note-I: The nomenclature for this project is as follows: primary, secondary and tertiary are named 'sectors', while agriculture, manufacturing etc. are named 'sub-sectors'.

Note-II: If the top 3 sub-sectors contribute to, say, 79% of the GDP of some category, you can report "These top 3 sub-sectors contribute to approximately 80% of the GDP". This is to simplify the analysis and make the results consumable. (Remember, the CEO has to present the report to the CMs, and CMs have limited time; so, the analysis needs to be sharp and concise.)

Plot the contribution of the sub-sectors as a percentage of the GSDP of each category.  

```python
# Get all the fields that belong to the c1 states

df_C1 = df_states.loc[df_states.State.isin(c1.State)&(df_states['S.No.']!='Total')&
        (~df_states['Item'].isin(['TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products',"Population ('00)",'Per Capita GSDP (Rs.)']))]

# Keeping only the necessary fields and grouping and sorting as required

df_C1 = df_C1[['Item','2014-15']].groupby(by='Item').sum().sort_values(by='2014-15',ascending=False).reset_index()

# Creating a new column Percentage_of_GSDP for easy analysis

df_C1['%_of_GSDP_Contribution'] = df_C1['2014-15']/(df_C1['2014-15'][0])*100 # here index index 0 has GSDP since we have sorted in descending order

# Finding which are the Sub-sectors that contribute approximately 80% to the GSDP (It should be 3 or more)

start =1; End = 4 # Takinfg first top 3 sectors initially to check whether it contributes approximately 80%. Starting with 1 to avoid first row which is GSDP

while df_C1.iloc[start:End ,-1].sum() <= 78: #considering anything less than or equal to 78% does not contribute 80% approximately, only equal to greater than 79% does.
    End = End+1
    
# Contribution of subsectors approximately 80% For category C1 to the total GSDP are as follows

C1_Sub_Sectors_contributes_80_percent_apprx = df_C1[['Item','%_of_GSDP_Contribution']].iloc[start:End].append({'Item':'ABOVE C1 SUB-SECTORS EXACT CONTRIBUTION =','%_of_GSDP_Contribution':round(df_C1.iloc[start:End ,-1].sum(),2)},ignore_index=True).rename(columns={'Item':'C1_Sub_Sectors_that_contributes_80%_approximately_to_GSDP_in_Total'})

C1_Sub_Sectors_contributes_80_percent_apprx
```

```python
plt.figure(figsize=(14,6))

# Selecting the required rows and columns and plotting the graph

top_subc1 = C1_Sub_Sectors_contributes_80_percent_apprx.set_index("C1_Sub_Sectors_that_contributes_80%_approximately_to_GSDP_in_Total").iloc[:-1,:]['%_of_GSDP_Contribution']

barplot_5 = sns.barplot(top_subc1.index,top_subc1.values)

plt.ylabel('Sub Sectors GSDP Percentage (%)',fontname="Arial",fontsize='15')

plt.xlabel('Sub-Sectors of C1',fontname="Arial",fontsize='15')

plt.title('Top Sub Sectors That Contributed 80% (approx) of the C1 GSDP. Exact Contribution is: {0}%'.format(C1_Sub_Sectors_contributes_80_percent_apprx.iloc[-1:,-1:].values[0][0]),fontname="Arial",fontsize='15',fontweight = 'bold')

plt.xticks(rotation = 90)

plt.show()
```

```python
# In above States other than union territories and the deleted states that had Null Values we have two states with wrong name (Uttrakhand and Chhatisgarh)
# We are going to correct the names to Uttarakhand and Chhattisgarh. If not while merging the columns with the part-1 df which had per-capita values, we will miss these 2 rows.

df_dropout = df_dropout.replace(['Chhatisgarh','Uttrakhand'],['Chhattisgarh','Uttarakhand'])

df_dropout
```

```python
# Merging two dataframes to get the per-capita-GSDP and the dropout rate.

drpout_percap = pd.merge(df_states[df_states.Item=='Per Capita GSDP (Rs.)'], df_dropout, how = 'inner', on = 'State')

drpout_percap
```

```python
# Adding a new column: 'Total_dropout_in_2014-15'

drpout_percap['Total_dropout_2014-15'] = drpout_percap.iloc[:,-3:].sum(axis = 1)

drpout_percap

```

```python
# Plotting agains Dropout rates and GSDP

#sns.set(style = "ticks")

x = drpout_percap['2014-15'].values # per-capita GSDP

y1 = drpout_percap['Primary - 2014-2015'].values # primary dropout

y2 = drpout_percap['Upper Primary - 2014-2015'].values # upper primary dropout

y3 = drpout_percap['Secondary - 2014-2015'].values # Secondary dropout

y4 = drpout_percap['Total_dropout_2014-15'].values # Total_dropout_in_2014-15

plt.figure(figsize=(14,12))

plt.subplot(221)

plt.title('GSDP vs Dropout Rate in Primary During 2014-2015',fontname="Arial",fontsize='15',fontweight='bold')

plt.xlabel('GSDP in Crores (Rs.)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.scatter(x,y1,color = 'blue')

plt.subplot(222)

plt.title('GSDP vs Dropout Rate in Upper-Primary During 2014-2015',fontname="Arial",fontsize='15',fontweight='bold')

plt.xlabel('GSDP in Crores (Rs.)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.scatter(x,y2,color = 'green')

plt.subplot(223)

plt.title('GSDP vs Dropout Rate in Secondary During 2014-2015',fontname="Arial",fontsize='15',fontweight='bold')

plt.xlabel('GSDP in Crores (Rs.)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.scatter(x,y3,color = 'orange')

plt.subplot(224)

plt.title('GSDP vs Total Dropout Rate During 2014-2015',fontname="Arial",fontsize='15',fontweight='bold')

plt.xlabel('GSDP in Crores (Rs.)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.scatter(x,y4,color = 'teal')

plt.show()
```

**Insight:** The above Scatter plot roughly shows that there is an increase in GSDP with decrease in Dropout rate. Let's try to get a Deep insight using the Linear regression model.

```python
# Function to viualise the Linear regression line

from numpy import *
import matplotlib.pyplot as plt

def plot_best_fit(intercept, slope):
    axes = plt.gca()
    x_vals = array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, 'r-')

# Utilising Linear Regression Algorithm from Sklearn

from sklearn.linear_model import LinearRegression
    
def Regression(X,Y):   
    regr = LinearRegression()
    regr.fit(X,Y)
    return regr

# Main Code To Plot the Graph

plt.figure(figsize=(14,12))

plt.subplot(221)

plt.scatter(x,y1,color = 'blue')

regr = Regression(x.reshape(-1,1) ,y1.reshape(-1,1))

plot_best_fit((regr.intercept_)[0],(regr.coef_)[0][0])

plt.title('GSDP vs Dropout Rate in Primary During 2014-2015',fontname="Arial",fontsize='15',fontweight = 'bold')

plt.xlabel('GSDP in Crores (Fig 1)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.subplot(222)

plt.scatter(x,y2,color = 'green')

regr = Regression(x.reshape(-1,1) ,y2.reshape(-1,1))

plot_best_fit((regr.intercept_)[0],(regr.coef_)[0][0])

plt.title('GSDP vs Dropout Rate in Upper-Primary During 2014-2015',fontname="Arial",fontsize='15',fontweight = 'bold')

plt.xlabel('GSDP in Crores (Fig 2)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.subplot(223)

plt.scatter(x,y3,color = 'orange')

regr = Regression(x.reshape(-1,1) ,y3.reshape(-1,1))

plot_best_fit((regr.intercept_)[0],(regr.coef_)[0][0])

plt.title('GSDP vs Dropout Rate in Secondary During 2014-2015',fontname="Arial",fontsize='15',fontweight = 'bold')

plt.xlabel('GSDP in Crores (Fig 3)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.subplot(224)

plt.scatter(x,y4,color = 'teal')

regr = Regression(x.reshape(-1,1) ,y4.reshape(-1,1))

plot_best_fit((regr.intercept_)[0],(regr.coef_)[0][0])

plt.title('GSDP vs Total Dropout Rate During 2014-2015',fontname="Arial",fontsize='15',fontweight = 'bold')

plt.xlabel('GSDP in Crores (Fig 4)',fontname="Arial",fontsize='15')

plt.ylabel('Dropout rate in Percentage',fontname="Arial",fontsize='15')

plt.show()
```

Comparing Figures 1&2 with Figures 3, the regression line is similar in figure3 comparing to figures 1&2. But the y scale in Figure3 is 2 times more.
